```python
# (class practice)project : bank account
def build():
    name = input('帳戶名稱: ')
    money = int(input('儲存金額: '))
    password = input('請輸入密碼: ')
    if name == '':
        print('無輸入名字!')
    elif money <= 0:
        print('金額不足!')
    elif password == '':
        print('未設定密碼')
    else:
        print('創帳成功!\n帳戶名為:',name,'\n帳戶金額:',money,'\n請牢記您的密碼!')
    acc = Account(name, money, password)
    return acc
class Account:
    fail = '驗證失敗!'
    def __init__(self, name, money, password):
        self.acc_name = name
        self.acc_money = money
        self.acc_pw = password
    def pw_check(self):
        check = input('密碼驗證: ')
        if check == self.acc_pw:
            print('密碼驗證成功!')
            return 1
        else:
            print('密碼輸入錯誤!')
            return 0
    def withdraw(self, take):
        if take <= self.acc_money:
            print('可以提領，請稍後...')
            self.acc_money -= take
            print('帳戶剩餘金額:',self.acc_money)
        else:
            print('金額不足，無法提領')
    def save(self, savemoney):
        print('帳戶金額:',self.acc_money)
        ch = input('確定執行?y/n: ')
        if ch == 'y':
            self.acc_money += savemoney
            print('存款成功，剩餘金額為:',self.acc_money)
        else:
            print('存款失敗')
while True:
    write = input('您的動作:')
    if write == 'build':
        acc = build()
    elif write == 'withdraw':
        c1 = acc.pw_check()
        if c1 == 1:
            take = int(input('我要提領:'))
            acc.withdraw(take)
        else:
            print(acc.fail)
    elif write == 'save':
        c2 = acc.pw_check()
        if c2 == 1:
            savemoney = int(input('我要存款: '))
            acc.save(savemoney)
        else:
            print(acc.fail)
    elif write == 'leave':
        print('多謝光臨，期待您的下次到來!')
        break
    else:
        print('未知指令!')
```
