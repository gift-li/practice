```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('./diamond.csv')
x, y = data['carats'], data['price']
x, y= np.array(x), np.array(y)
x = x.reshape(23, 1)
y = y.reshape(23, 1)
```

```python
regr = LinearRegression()
regr.fit(x, y)
Y = regr.predict(x)
plt.scatter(x, y)
plt.plot(x, Y, 'r')
plt.show()
```
